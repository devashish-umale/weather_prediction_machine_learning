```python
for i in num:
    df[i].fillna(df[i].median(), inplace = True)
```

```python
for i in cat:
    df[i].fillna(df[i].mode()[0], inplace = True)
```

```python
import seaborn as sns
```

#outlier Treatment

```python
for i in num:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3 - q1
    lf = q1 - 1.5* iqr
    uf = q3 + 1.5 * iqr
    condition = (df[i] < lf) | (df[i] > uf)
    df.loc[condition, i] = df[i].median()
```

```python
from sklearn.feature_selection import VarianceThreshold
threshold = 0.05
selector = VarianceThreshold(threshold = threshold)
X_filtered = selector.fit_transform(X)
selected = selector.get_support(indices = True)
selections = list(X.columns[selected])
inputs = X[selections]
inputs
```

```python
X = inputs.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
def model_performance(actual, predicted):
    print("accuracy-score: ", accuracy_score(actual, predicted) )
    print("confusion_matrix: ")
    print(confusion_matrix(actual, predicted))
    print("Classification_report : ")
    print(classification_report(actual, predicted))
    print("roc_auc_score :", roc_auc_score(actual, predicted))
```

```python
model_performance(y_pred, y_train)
```

```python
y_test_pred = lr.predict(X_test)
```

```python
model_performance(y_test_pred,y_test)
```
